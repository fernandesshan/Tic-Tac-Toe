```python
# Create display for board

from IPython.display import clear_output
def display_board(board):
    clear_output()
    print("    |     |   ")
    print(f' {board[7]}  |  {board[8]}  |  {board[9]} ')
    print("    |     |   ")
    print("__ __ __ __ __ _")
    print("                ")
    print("    |     |   ")
    print(f' {board[4]}  |  {board[5]}  |  {board[6]} ')
    print("    |     |   ")
    print("__ __ __ __ __ _")
    print("                ")
    print("    |     |   ")
    print(f' {board[1]}  |  {board[2]}  |  {board[3]} ')
    print("    |     |   ")
    
    
# Let player1 choose X or O
def X_or_O():
    marker=""
    while marker not in ('X','O'):
        marker = input('Player 1, do you want X or O? ').upper()
    if marker=='X':
        return ('X','O')
    else:
        return ('O','X')
    
# Randomly select which player will go first
import random

def choose_first():
    player = random.randint(1, 2)
    print(f'player {player} plays first')
    return player

# Check if position for marking is free
def space_check(board, position):
    return board[position] == ' '

# Take position of marker from user
# Check if number is betwwen (1-9) and if space is empty
def player_choice(board, player):
    pos = 0
    acceptable_lst = [1,2,3,4,5,6,7,8,9]
    
    while (pos not in acceptable_lst) or not space_check(board, pos):
        try:
            pos = int(input(f'Player {player} enter the position {acceptable_lst} where you want to place your mark: '))
        
        except ValueError:
            print("Wrong Input!!! Try Again")
            continue

        if(pos not in acceptable_lst):
            print(f'Incorrect option. Please select from {acceptable_lst}')
        elif( not space_check(board,pos)):
            print("This space is already filled")
        else:
            return pos
        
# Place marker on board
def place_marker(board, marker, position):
    board[position]= marker
#     return board

# Check if player won
def win_check(board, mark):
    if board[1]==board[2]==board[3]==mark or board[4]==board[5]==board[6]==mark or board[7]==board[8]==board[9]==mark or board[1]==board[5]==board[9]==mark or board[3]==board[5]==board[7]==mark or board[1]==board[4]==board[7]==mark or board[2]==board[5]==board[8]==mark or board[3]==board[6]==board[9]==mark:
        print("You won !")
        return True
    else:
        return False
    
# Check if board is full
def full_board_check(board):
    for space in board:
        if space == " ":
#             print("Full board checked - False")
            return False
#     print("Full board checked - True")
    return True

# Check if plays wants to play again
def replay():
    play=""
    while play not in ('Y','y','N','n'):
        
        play = input("Do you want to play again? Y or N: ")
        if play == 'Y' or play =='y':
            return True
        elif play =='N' or play =='n':
            return False
        else:
            print("Invalid input. Please try again")
```

```python
# Game setup
print('Welcome to Tic Tac Toe!')
demo_board=['#','1','2','3','4','5','6','7','8','9' ]
display_board(demo_board)
print("Select board position as mentioned above")


while True:
    final_board=['#',' ',' ',' ',' ',' ',' ',' ',' ',' ']
    
    #Set up board > Choose markers >Decide who plays first
    (player1_marker, player2_marker) = X_or_O()
    
    player = choose_first()

    play_game = ''
    while play_game not in ('y','Y','n','N','Yes','No'):
        play_game = input('Are you ready to play? Enter Yes or No.')
        if play_game.lower() == 'y' or  play_game == 'Yes':
            game_on = True
        elif play_game.lower() == 'n' or  play_game == 'No':
            game_on = False
            
        else:
            print("Invalid input. Please say Y or N")

    while game_on:
        if player == 1:
            #Show the board
            display_board(final_board)
            
            #Choose marker position
            pos1= player_choice(final_board, player)
            
            #Place the marker
            place_marker(final_board, player1_marker, pos1)

            #Check if won
            if win_check(final_board, player1_marker):
                display_board(final_board)
                print("Player 1 wins")
                game_on= False
                
            #Check if tie
            else:
                if full_board_check(final_board):
                    display_board(final_board)
                    print("It's a tie !")
                    break
                    
                #Next players turn
                else:
                    player = 2
        elif player == 2:
            #Show the board
            display_board(final_board)
            
            #Choose marker position
            pos2= player_choice(final_board, player)
            
            #Place the marker
            place_marker(final_board, player2_marker, pos2)

            #Check if won
            if(win_check(final_board, player2_marker)):
                display_board(final_board)
                print("Player 2 wins")
                game_on= False
                
            #Check if tie
            else:
                if(full_board_check(final_board)):
                    display_board(final_board)
                    print("It's a tie !")
                    break
                    
                #Next players turn
                else:
                    player = 1


    if not replay():
        break
```
